# Задание по методам строк и логическим выражениям


### **Проверка корректности записей в журнале пациентов**

Вы ведёте журнал приёма пациентов с туберлёзом. Каждая запись должна соответствовать строгому формату, чтобы её можно было автоматически обрабатывать:

**ФАМИЛИЯ;ИМЯ;ДАТА_РОЖДЕНИЯ;ДИАГНОЗ;СТАТУС**

---

Пример корректной записи:

**Иванов;Анна;1985-03-22;A15.0;активный**

---

**Правила формата:**
1. Фамилия и имя — только буквы (без пробелов, цифр и символов).
2. Дата рождения — в формате `ГГГГ-ММ-ДД`.
3. Диагноз — по МКБ-10: начинается с буквы `A`, `B` или `C`, затем 2–3 цифры + может быть точка с ещё одной цифрой (например, `A15.0`, `B20`, `C01.1`).
4. Статус — одно из: `активный`, `ремиссия`, `выписан`.

---

### Само задание:
Напишите скрипт который
- Принимает строку — одну запись из журнала.
- Возвращает `True`, если запись **корректна**, и `False` — если нарушены любые из правил.

**Используйте:**
- Методы строк: `.split()`, `.isdigit()`, `.isalpha()`, `.startswith()`, `.replace()` и др.
- Логические выражения (`and`, `or`, `not`).
- Проверку длины подстрок и состава символов.

---

### **Примеры:**

- "Петров;Сергей;1970-11-05;A15.0;активный" - # True
- "Сидорова;Мария1;1990-01-30;B20;ремиссия" - # False (в имени цифра)
- "Кузнецов;Дмитрий;1988-13-01;A16;выписан" - # False (некорректный месяц)
- "Васильева;Елена;2000-02-29;C01.1;наблюдение" - # False (неверный статус)
```

Если какие-то методы строк вам не ясны, смело изучайте их через документацию
