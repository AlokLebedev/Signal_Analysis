```python
# Create database
analysis_history = (
 ("Общий_анализ_крови_лейкоциты", 20240315, 9.8),
 ("Мочевина", 20240315, 6.2),
 ("Креатинин", 20240322, 88),
 ("Общий_анализ_крови_лейкоциты", 20240322, 7.3),
 ("Общий_анализ_крови_лейкоциты", 20250426, 7.1),
 ("Общий_анализ_крови_лейкоциты", 20250526, 7.1),
 ("Общий_анализ_крови_лейкоциты", 20260426, 3.1))
```

```python
# вытаскиваем из "кортежа кортежей" нужные нам кортежи (с показателями лейкоцитов)
analysis_leukocytes = []
for k in analysis_history: 
    if k[0] == "Общий_анализ_крови_лейкоциты":
        analysis_leukocytes.append(k)
    analysis_leukocytes.sort()
# делаем цикл, чтобы понять как изменился показатель
differents_between_value = -1
for leukocytes in analysis_leukocytes: 
    leukocytes_str = str(leukocytes[1])  
    if differents_between_value == -1:
        print (f"Дата {leukocytes_str[:4]}.{leukocytes_str[4:6]}.{leukocytes_str[6:8]}, Количество лекоцитов {leukocytes[2]}.показатель новый")
    elif ((leukocytes[2]-6.5)**2>differents_between_value):
        print (f"Дата {leukocytes_str[:4]}.{leukocytes_str[4:6]}.{leukocytes_str[6:8]}, Количество лекоцитов {leukocytes[2]}.показатель ухудшился")
    elif ((leukocytes[2]-6.5)**2==differents_between_value):
        print (f"Дата {leukocytes_str[:4]}.{leukocytes_str[4:6]}.{leukocytes_str[6:8]}, Количество лекоцитов {leukocytes[2]}.показатель не изменился")
    else:
        print (f"Дата {leukocytes_str[:4]}.{leukocytes_str[4:6]}.{leukocytes_str[6:8]}, Количество лекоцитов {leukocytes[2]}.показатель улучшился")
    differents_between_value = (leukocytes[2]-6.5)**2
print(*analysis_leukocytes, sep="\n") # заюзал конструкцию с (*): print(*my_list, sep="\n")
```
