```python
temperatures = [36.6, 36.7, 36.8, 36.9, 37.5, 37.2, 37.0, 37.0, 37.9, 37.0, 36.9, 36.8]
```

```python
# I find max temperature
danger = 0
for t in temperatures:
    if t >= 37.1:
        danger = danger + 1
    else:
        danger = danger + 0

if danger >= 1:
    print('внимание')
else:
    print('все ок')

```

```python
# I find time when temp was max (and learn f-strings)
max_t = max(temperatures)
n = temperatures.index(max_t)*2
print (f"Максимум {max_t} °C в {n:02}.00" )
```

```python
# I find how many values were increased
increases = 0
for t in temperatures:
    if t >= 37.0:
        increases +=1
    else:
        increases +=0
print (increases)
```

```python
pressure_group1 = [120, 127, 125, 130, 135, 132, 128]
pressure_group2 = [140, 142, 145, 150, 138, 143, 147, 160]
```

```python
# здесь решил прописать фулл подробно с точки зрения математики, возился долго, но было интересно почувствовать себя бестолковым)
#  find average
average_1 = sum(pressure_group1)/len(pressure_group1)
average_2 = sum(pressure_group2)/len(pressure_group2)

#find mediane
pressure_group1.sort()
i_1=len(pressure_group1)
if i_1 % 2 == 0:
    mediane_1 = (pressure_group1 [i_1//2] + pressure_group1 [i_1//2 - 1])//2
else: 
    mediane_1 = pressure_group1 [int(i_1/2)]


pressure_group2.sort()
i_2=len(pressure_group2)
if i_2 % 2 == 0:
    mediane_2 = (pressure_group2 [i_2//2] +pressure_group2 [i_2//2 - 1])/2
else: 
    mediane_2 = pressure_group2 [int(i_2/2)]
 



print(f"average1= {average_1}")
print(f"average2= {average_2}")
print(f"mediane_1= {mediane_1}")
print(f"mediane_2= {mediane_2}")
```

```python

# считаем стандартное отклонение и дисперсию
differences_squares = [(p - average_1) ** 2 for p in pressure_group1]

standard_deviation = sum(differences_squares)/(len(pressure_group1)-1)
variance = standard_deviation**0.5

print(standard_deviation)
print(variance)

differences_squares2 = [(p - average_2) ** 2 for p in pressure_group2]

standard_deviation2 = sum(differences_squares2)/(len(pressure_group2)-1)
variance2 = standard_deviation2**0.5

print(standard_deviation2)
print(variance2)
```
